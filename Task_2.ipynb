# **Upload Dataset**

```python
import pandas as pd

df = pd.read_csv('train.csv')
```

# **Data Cleaning**

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
df['HasCabin'] = df['Cabin'].notnull().astype(int)
df.drop(columns=['Cabin'], inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Pclass'] = df['Pclass'].astype(str)
```

# **Statistics & Group Insights**

```python
df.describe(include='all')
```

```python
df.groupby('Sex')['Survived'].mean()
df.groupby('Pclass')['Survived'].mean()
df.pivot_table(index='Sex', columns='Pclass', values='Survived')
```

# **Data Visualization**

**Correlation Heatmap**

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

**Survival by Gender**

```python
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Survival Count by Gender')
plt.show()
```

**Survival Count by Class**

```python
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Survival Count by Class')
plt.show()
```

**Age Distribution by Survival**

```python
sns.histplot(data=df, x='Age', hue='Survived', multiple='stack')
plt.title('Age Distribution by Survival')
plt.show()
```

**Survival Rate by Class and Gender**

```python
sns.barplot(x='Pclass', y='Survived', hue='Sex', data=df)
plt.title('Survival Rate by Class and Gender')
plt.show()
```

**Survival Rate by Gender and Embarked Port**

```python
pivot = df.pivot_table(index='Sex', columns='Embarked', values='Survived')
sns.heatmap(pivot, annot=True, cmap='YlGnBu')
plt.title('Survival Rate by Gender and Embarked Port')
plt.show()
```
